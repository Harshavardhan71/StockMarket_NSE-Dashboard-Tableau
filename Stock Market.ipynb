```python
adani = pd.read_csv('Source/ADANIPORTS.csv')
airtel = pd.read_csv('Source/BHARTIARTL.csv')
hdfc = pd.read_csv("Source/HDFCBANK.csv")
maruti = pd.read_csv("Source/MARUTI.csv")
reliance = pd.read_csv('Source/RELIANCE.csv')
tcs = pd.read_csv('Source/TCS.csv')
```

```python
dfs = [adani, airtel, hdfc, maruti, reliance, tcs]
names = ['adani', 'airtel', 'hdfc', 'maruti', 'reliance', 'tcs']
```

```python
for i, df in enumerate(dfs):
    # Selecting and renaming columns
    df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Last', 'Volume']]
    df = df.rename(columns={'Last': 'Adj Close'})

    # Adding moving averages
    df['MA50'] = df['Close'].rolling(50).mean()
    df['MA200'] = df['Close'].rolling(200).mean()

    # Calculating price and volume changes
    df['Previous day close price'] = df['Close'].shift(1)
    df['Change in price'] = df['Close'] - df['Previous day close price']
    df['Percent change in price'] = df['Close'].pct_change()
    df['Previous day volume'] = df['Volume'].shift(1)
    df['Change in volume'] = df['Volume'] - df['Previous day volume']
    df['Percent change in volume'] = df['Volume'].pct_change()

    # Print first few rows of the cleaned DataFrame
    print(f"Head of {names[i]}:")
    print(df.head())  # Print first 5 rows of the DataFrame
    print("\n")  # Add a new line between outputs

    # Save each DataFrame to a CSV file
    df.to_csv(f'cleaned_data/{names[i]}.csv', index=False)
```
